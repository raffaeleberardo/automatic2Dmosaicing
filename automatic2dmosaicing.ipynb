## Create frame function
We use this function **to create a new black frame** where we paste the reference image and then the warped image with the greater number of correspondencies. We do this augmenting the border of the reference image with the dimensions of the image that we will use to create the panorama. In this way after we will do the stitching without the knowledge of the image position in the frame **(we don't know if we will put the image on the left, right, top or bottom, and it isn't of our concern)**. 
We have to do this otherwise we would go through an error because **we can't just perform the warping without knowing the frame size in which we perform this operation**.

```python
def createFrame(ref_img, img):
    '''

    Make a border around the ref_img equal to the shape of img. We use this as a frame where we paste the warped img.

    :param ref_img: reference image
    :param img: image with which we want to make the stitching
    :return: new ref_img
    '''
    top = int(img.shape[0])
    bottom = top
    right = int(img.shape[1])
    left = right

    ref_img = cv.copyMakeBorder(ref_img, top, bottom, left, right, cv.BORDER_CONSTANT, None, 0)
    return ref_img
```

## Cropping function
The cropping function peform an adjustment of the frame by removing the extra black background that you can see in the intermediate steps of the stitching procedure.

```python
def cropping(img):
    '''

    :param img: image where we delete the black background
    :return: image without black background
    '''
    gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    _, thresh = cv.threshold(gray, 1, 255, cv.THRESH_BINARY)
    contours, hierarchy = cv.findContours(thresh, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
    cnt = contours[0]
    x, y, w, h = cv.boundingRect(cnt)
    crop = img[y:y + h, x:x + w]
    return crop
```

## Good matches function
We use a k-nearest neighbor algorithm to find 2 matches with a point in the reference image and then we use the Lowe's ratio test to discriminate one of the two matches in order to have a correspondence one-to-one between our points.

To filter the matches, Lowe proposed in to use a distance ratio test to try to eliminate false matches. The distance ratio between the two nearest matches of a considered keypoint is computed and it is a good match when this value is below a threshold. Indeed, this ratio allows helping to discriminate between ambiguous matches (distance ratio between the two nearest neighbors is close to one) and well discriminated matches. 

```python
def goodMatches(ref_des, des, matcher):
    '''

    :param ref_des: reference descriptors of reference image ref_img
    :param des: descriptors of image img
    :param matcher: matcher use to compute near points between ref_img and img
    :return: good found matches
    '''
    # compute the matches and keep 2 nearest neighbour
    matches = matcher.knnMatch(des, ref_des, k=2)

    # store all the good matches as per Lowe's ratio test.
    good = []
    for m, n in matches:
        if m.distance < 0.7 * n.distance:
            good.append(m)
    return good
```

## Color correction

In whatever way you choose to merge the images, you don't have to believe that the color with remain uniform in the result, because the camera lens typycally give more brightness to the center and less when you move further from it, so in the result you probably will see a color difference.

In our case we found an interesting method, very simple but effective, that we could apply from scratch. The method consists in computing two weights for our images and the apply a weightened sum of these.

This is the formula:
$$
    resultingImage = \frac{reference\_image * w1 + warped\_image * w2}{w1 + w2}
$$

In which the weights are computed using the *distance_transform_edt()* function of the python *scipy* library.
This function gives more weight to the pixel at the center of the image and less to the pixel closer to the edge of the image and it turns out that with this operation we obtain a uniform color in the image result.

```python
def blending(I1, I2):

    '''

    :param I1: first image with which we do the blending
    :param I2: second image with which we do the blending
    :return: Blended image

     WEIGHTENED BLENDING:
     I_blended = (I1 * w1 + I2 * w2)/(w1 + w2)

     For us I1 = ref_img, I2 = warped_img
     distance_transform_edt gives more weight to the px in the centre of the image and less on the borders.
     Link: https://www.youtube.com/watch?v=D9rAOAL12SY

    '''

    w1 = distance_transform_edt(I1) # This command correspond to the bwdist() in MATLAB
    w1 = np.divide(w1, np.max(w1))
    w2 = distance_transform_edt(I2)
    w2 = np.divide(w2, np.max(w2))

    # STITCHING THE TWO IMAGES
    I_blended = np.add(np.multiply(I1, w1), np.multiply(I2, w2))
    w_tot = w1 + w2
    I_blended = np.divide(I_blended, w_tot, out=np.zeros_like(I_blended), where=w_tot != 0).astype("uint8")
    return I_blended
```

## SIFT
**SIFT (Scale Invariant Feature Transform)** is a feature detection algorithm used in Computer Vision which helps in finding the local features in an image, commonly know as *keypoints* of the image.
These keypoints are *scale and rotation invariant* and this is useful in different computer vision applications as *image matching, object detection, scene detection etc...*
The important thing is that **the SIFT features are not affected by the size or orientation of the image**, so we can have matches also between images of different size and orientation.
The entire process to detect these keypoints can be divided in 4 parts:
1. **Constructing a Scale Space** to make sure that features are scale independent.
2. **Keypoints Localization** to identify the suitable features
3. **Orientation Assignment** to ensure that the keypoints are *rotation invariant*
4. **Keypoint Descriptor** to assign a unique fingerprint to each keypoint

**In our case, we use these keypoints for feature matching between our images set**

### Constructing the Scale Space

We need to identify the most distinct features in a given image while ignoring any noise. Additionally, we need to ensure that the features are not scale-dependent.
To reduce the noise in the image we used the **Gaussian Blurring technique**. So, for every pixel in an image, the Gaussian Blur calculates a value based on its neighboring pixels. The results is that the texture and minor details are removed from the image and only the relevant information (like the shape and edges) remain.
Once we did this we need to ensure that these features **must not be scale-dependent**. This means we will be searching for these features on multiple scales, by creating the so called **scale space**, a collection of images having different scales, generated from a single image. Hence, the blur images are created for multiple scales. To crate a new set of images of different scales, we will take the original imae and reduce tha scale by half. For each new image, we will create some blur versions.
How many times do we need to scale the image and how many subsequent blur images need to be created for each scaled image? The **ideal number of octaves should be four, and for each octave, the number of blur images should be five.**

#### Difference of Gaussian
We try to enhance the features using a technique called **Difference of Gaussian**, a feature enhancement algorithm that involves the subtraction of the one blurred version of an original image from another, less blurred version of the original.
So the DoG creates another set of images, for each octave, by subtracting every image from the previous image in the same scale. We have to implement this process for all the octaves (scaled image size) that we created.

### Keypoint Localization
Once the image have been created, the next step is to find the important keypoints from the image that can be used for feature matching. **The idea is to find the local maxima and minima for the images**. This part is divided into 2 steps:
1. Find the local maxima and minima
2. Remove low contrast keypoints (keypoint selection)

#### Local Maxima and Minima
To localte the local maxima and minima, we go through every pixel in the image and compare it with its neighboring pixels.
With neighboring pixels we include the surrounding pixels of the considered image and the nine pixels for the previous and next image in the octave.
This means that every pixel value is compared with 26 other pixel values to find whether it is the local maxima/minima. 
With this method we obtain the potential keypoints to ensure that these are the most accurate keypoints to represent the image.

<img src="notebook_images/localmaxmin.png"/>

#### Keypoint Selection

So far we have generated scale-invariant keypoints. But some of these keypoints may not be robust to noise. This is why we need to perform a final check to make sure that we have the most accurate keypoints to represent the image features.
**To do this we will eliminate the keypoints that have low contrast, or lie very close to the edge.**

To deal with the low contrast keypoints, a second order Taylor expansion is computed for each keypoint. If the resulting value is less than 0.03 in magnitude (pixel intensity), we reject the keypoint.
About the remaining keypoints we perform a check to identify the poorly located keypoints that are the keypoints close to the edge and that have a high edge response but may not be robust to a small amount of noise. A second order Hessian matrix is used to identify such keypoints.

Now that we have performed both the contrast test and the edge test to reject the unstable keypoints, we will now assign an orientation value for each keypoint **to make the rotation invariant.**

## Orientation Assignment

We will now assign an orientation to each of these keypoints so that they are invariant to rotation. We can again divide this step into two smaller steps:
1. Calculate the magnitude and orientation
2. Create a histogram for magnitude and orietation

### Calculate Magnitude and Orientation

To find the magnitude and orientation we have to compute the **gradients** in x and y direction for the neighboring of the considered pixel in an image.
Once we have the gradients we can find the magnitude and orientation using the following formulas:

$ Magnitude = \sqrt{G_x^2 + G_y^2} $

$ \phi = \arctan{ \frac{G_y}{G_x} } $

**The Magnitude represents the intensity of the pixel and the orientation gives the direction for the same.**
We can now create a histogram given that we have these magnitude and orientation values for the pixels.

### Creating a Histogram for Magnitude and Orientation

On the x-axis, we will have bins for angle values, like 0-9, 10-19, up to 360. On the y-axis we will have the magnitude value for the given angle. We will classify all the pixels around the keypoints.
The histogram woul peak at some point. **The bin at which we see the peak will be the orientation for the keypoint.** Additionally, if there is another significant peak (seen between 80 - 100%), the another keypoint is generated with the magnitude and scale the same keypoiint used to generate the histogram. And the angle or orientation will be equal to the new bin that has the peak.
Effectively at this point, we can say that there can be a small increase in the number of keypoints.

## Keypoint Descriptor

This is the final step for SIFT. So far, we have stable keypoints that are scale-invariant and rotation invariant. In this section, we will use the neighboring pixels, their orientations, and magnitude, to generate a unique fingerprint for this keypoint. This fingerprint is called **descriptor**.

Additionally, since we use the surrounding pixels, the descriptors will be partially invariant to illumination or brighteness of the images.
We will first take a 16x16 neighborhood around the keypoint. This 16x16 block is further divided into 4x4 sub-blocks and for each of these sub-blocks, we generate the histogram using a magnitude and orientation.

At this stage, the bin size is increased and we take only 8 bins. Each of these arrows represents the 8 bins and the length of the arrows define the magnitude. So, we will have a total of 128 bn values for every keypoint.

## Flann Matcher

FLANN (Fast Library for Approximate Nearest Neighbors) is a library for performing fast approximate nearest neighbor searches in high dimensional spaces. It contains a collection of algorithms we found to work best for nearest neighbor search.

We choose a knn algorithm to search correspondencies between two images.

```python
# To compute the matches between pair of image we used a k-tree algorithm available in the OpenCV library 
FLANN_INDEX_KDTREE = 1
index_params = dict(algorithm=FLANN_INDEX_KDTREE, trees=2)

# If we set an higher parameter for the checks we should have more precision in searching correspondencies,
# but more computational costs

search_params = dict() #  dict(checks=500)
flann = cv.FlannBasedMatcher(index_params, search_params)

# we want a minimum of 4 number of matching points to estimate the homography
# we want also enough points to discriminate an image from one that isn't a representation of the same scene, this is why we choose a greater number than 4,
# 50 in our case.

MIN_MATCH_COUNT = 50
```

## Main steps for the creation of the panorama
We start to do this by searching the image with the greater number of correspondencies with our reference, using the little function explained above (best_parameters()). Then, if the number of correspondencies is greater than a chosen threshold (we want to have a minimum number of correspondencies) we start the computation of the homography matrix and then we perform the warping operation between the reference image and the chosen image, using the found homography. Then we perform the stitching and we correct the color of the obtained panorama (we want to have an uniform color for our panorama, without seeing the points where the two images overlap).
Then we reiterate this procedure until:

- we have no more images in the list
- we don't have a sufficient number of correspondencies

We took inspiration for this procedure by using the algorithm of the course book (*Multiple View Geometry for Computer Vision - page 206-207*)

> This is the algorithm:
> 1. Choose one image of the set as a reference.
> 2. Compute the homography H which maps one of the other images of the set to this reference image.
> 3. Projectively warp the image with this homography, and augment the reference image with the non-overlapping part of the  warped image.
> 4. Repeat the last two steps for the remaining images of the set.
> 
> The homographies may be computed by identifying (at least) four corresponding
> points, or by using the automatic method of algorithm.

Now we will explain in detail some parts of this procedure.

### What is the homography matrix?
In simple words, an **homography H maps images of points which lie on a world plane from one camera view to another**.
If we find correspondences between some points of our planes, H maps points of the first plane to the correspondencies on the second plane.
So in our project we have to use this matrix after we found correspondencies between our reference image and another image from the remained on the set.
We compute the homography matrix using the OpenCV function *findHomgraphy()* that takes as parameters the source points (src_pts), the destination points (dst_pts), the method used to compute the homography (in our case we choose the RANSAC method, explained later) and the threshold used by the RANSAC method.
To be clear, the source points identify the keypoints of the image that has to be stitched with the reference and the destination points identify the keypoints for the reference image. In other words, the source points represent where i am and the destination points where I have to go.
Now we explain a little bit how the RANSAC method works to extimate our homography in a robust way.

### RANSAC method

The method RANSAC tries many different random subsets of the corresponding point pairs (of four pairs each, collinear pairs are discarded), estimate the homography matrix using this subset and a simple least-squares algorithm, and then compute the quality/goodness of the computed homography (which is the number of inliers for RANSAC). The best subset is then used to produce the initial estimate of the homography matrix and the mask of inliers/outliers.

Regardless of the method, robust or not, the computed homography matrix is refined further (using inliers only in case of a robust method) with the Levenberg-Marquardt method to reduce the re-projection error even more.

The methods RANSAC can handle practically any ratio of outliers but need a threshold to distinguish inliers from outliers.


### Warping operation
The *warpPerspective()* function applies a perspective transformation to an image, in our case we transform the image with the greater number of corresponding points in order to match the reference image and then we add the two to obtain the stitched image (with no color correction).

The function transform the source image using the specified matrix, in our case the Homography H according to this formula.

$$
dst(x, y) = src\left(\frac{M_{11}x + M_{12}y + M_{13}}{M_{31}x + M_{32}y + M_{33}}, \frac{M_{21}x + M_{22}y + M_{23}}{M_{31}x + M_{32}y + M_{33}}\right)
$$
